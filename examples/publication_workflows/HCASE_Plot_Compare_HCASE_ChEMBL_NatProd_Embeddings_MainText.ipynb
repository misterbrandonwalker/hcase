```python

import pandas as pd


import matplotlib
#matplotlib.use('agg')
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Config section



fname_embedding_canvass_natprod = '../data/canvass_emb_hcase_natprod_bms_dim_2.tab'
fname_embedding_drugs_natprod = '../data/drugs_emb_hcase_natprod_bms_dim_2.tab'

fname_embedding_canvass_chembl = '../data/canvass_emb_hcase_chembl_24_1_bms_dim_2.tab'
fname_embedding_drugs_chembl = '../data/drugs_emb_hcase_chembl.tab'

fname_out_compare_hcase_embeddings_plot = '../plots/comparative_embeddings/main_comparative_canvass_drugs_hcase_natprod.png'

```

```python
# Import datasets

df_embedded_canvass_natprod = pd.read_csv (fname_embedding_canvass_natprod, sep = '\t')
df_embedded_drugs_natprod = pd.read_csv (fname_embedding_drugs_natprod, sep = '\t')

df_embedded_canvass_chembl  = pd.read_csv (fname_embedding_canvass_chembl, sep = '\t')
df_embedded_drugs_chembl  = pd.read_csv (fname_embedding_drugs_chembl, sep = '\t')
```

```python
# Functions

def panel_dual_plot (subfig, df1, df2, subfig_label):



    subfig.scatter(df1['Dim_1'].values, df1['Dim_2'].values, c = 'blue', alpha = 0.3, marker = 'o', s = 20)
    subfig.scatter(df2['Dim_1'].values, df2['Dim_2'].values, c = 'yellow', alpha = 0.3, marker = 'o', s = 20)


    subfig.set_title (subfig_label, fontsize=24)
    subfig.set_xlabel ('Dim 1', fontsize=20)
    subfig.set_ylabel ('Dim 2', fontsize=20)
    
    return(subfig) 
```

```python
# Pre-process input data




df_8_canvass_chembl = df_embedded_canvass_chembl[df_embedded_canvass_chembl['hc_order'] == 8].copy()




df_8_drugs_chembl = df_embedded_drugs_chembl[df_embedded_drugs_chembl['hc_order'] == 8].copy()




df_5_canvass_natprod = df_embedded_canvass_natprod[df_embedded_canvass_natprod['hc_order'] == 5].copy()



df_5_drugs_natprod = df_embedded_drugs_natprod[df_embedded_drugs_natprod['hc_order'] == 5].copy()





```

```python
# Plotting
```

```python
# Compare HCASE ChEMBL Embeddings


fig = plt.figure()

fig, subfigures = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 16))




panel_dual_plot (subfigures[0], df_5_canvass_natprod, df_5_drugs_natprod, subfig_label = 'HCASE NatProd Space, z = 5')
panel_dual_plot (subfigures[1], df_8_canvass_chembl, df_8_canvass_chembl, subfig_label = 'HCASE ChEMBL Space, z = 8')


#subfigures[0][1].axis('off')
#subfigures[1][1].axis('off')




subfigures[0].text(-0.25, 1.25, "A", transform=subfigures[0].transAxes,
            fontsize=64, fontweight='bold', va='top')

subfigures[1].text(-0.25, 1.25, "B", transform=subfigures[1].transAxes,
            fontsize=64, fontweight='bold', va='top')

fig.suptitle('Comparison of Embeddings\nin HCASE ChEMBL and NatProd Spaces', fontsize = 32, y = 1.1)
#fig.supylabel('Dim 2')


plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.5)

plt.savefig (fname_out_compare_hcase_embeddings_plot, dpi=300)

plt.show()


```
