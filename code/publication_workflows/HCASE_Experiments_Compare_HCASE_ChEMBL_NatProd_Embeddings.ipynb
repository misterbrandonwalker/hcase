```python

import pandas as pd


import matplotlib
#matplotlib.use('agg')
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Config section



fname_embedding_canvass_natprod = '../data/canvass_emb_hcase_natprod_bms_dim_2.tab'
fname_embedding_drugs_natprod = '../data/drugs_emb_hcase_natprod_bms_dim_2.tab'

fname_embedding_canvass_chembl = '../data/canvass_emb_hcase_chembl_24_1_bms_dim_2.tab'
fname_embedding_drugs_chembl = '../data/drugs_emb_hcase_chembl.tab'

fname_out_compare_hcase_chembl_embedding_plot = '../plots/comparative_embeddings/comparative_canvass_drugs_hcase_chembl.png'
fname_out_compare_hcase_natprod_embedding_plot = '../plots/comparative_embeddings/comparative_canvass_drugs_hcase_natprod.png'

```

```python
# Import datasets

df_embedded_canvass_natprod = pd.read_csv (fname_embedding_canvass_natprod, sep = '\t')
df_embedded_drugs_natprod = pd.read_csv (fname_embedding_drugs_natprod, sep = '\t')

df_embedded_canvass_chembl  = pd.read_csv (fname_embedding_canvass_chembl, sep = '\t')
df_embedded_drugs_chembl  = pd.read_csv (fname_embedding_drugs_chembl, sep = '\t')
```

```python
# Functions

def panel_dual_plot (subfig, df1, df2, subfig_label):



    subfig.scatter(df1['Dim_1'].values, df1['Dim_2'].values, c = 'blue', alpha = 0.3, marker = 'o', s = 10)
    subfig.scatter(df2['Dim_1'].values, df2['Dim_2'].values, c = 'yellow', alpha = 0.3, marker = 'o', s = 10)


    subfig.set_title (subfig_label, fontsize=24)
    subfig.set_xlabel ('Dim 1', fontsize=20)
    subfig.set_ylabel ('Dim 2', fontsize=20)
    
    return(subfig) 
```

```python
# Pre-process input data


df_2_canvass_chembl = df_embedded_canvass_chembl[df_embedded_canvass_chembl['hc_order'] == 2].copy()
df_3_canvass_chembl = df_embedded_canvass_chembl[df_embedded_canvass_chembl['hc_order'] == 3].copy()
df_4_canvass_chembl = df_embedded_canvass_chembl[df_embedded_canvass_chembl['hc_order'] == 4].copy()
df_5_canvass_chembl = df_embedded_canvass_chembl[df_embedded_canvass_chembl['hc_order'] == 5].copy()
df_6_canvass_chembl = df_embedded_canvass_chembl[df_embedded_canvass_chembl['hc_order'] == 6].copy()
df_7_canvass_chembl = df_embedded_canvass_chembl[df_embedded_canvass_chembl['hc_order'] == 7].copy()
df_8_canvass_chembl = df_embedded_canvass_chembl[df_embedded_canvass_chembl['hc_order'] == 8].copy()

df_2_drugs_chembl = df_embedded_drugs_chembl[df_embedded_drugs_chembl['hc_order'] == 2].copy()
df_3_drugs_chembl = df_embedded_drugs_chembl[df_embedded_drugs_chembl['hc_order'] == 3].copy()
df_4_drugs_chembl = df_embedded_drugs_chembl[df_embedded_drugs_chembl['hc_order'] == 4].copy()
df_5_drugs_chembl = df_embedded_drugs_chembl[df_embedded_drugs_chembl['hc_order'] == 5].copy()
df_6_drugs_chembl = df_embedded_drugs_chembl[df_embedded_drugs_chembl['hc_order'] == 6].copy()
df_7_drugs_chembl = df_embedded_drugs_chembl[df_embedded_drugs_chembl['hc_order'] == 7].copy()
df_8_drugs_chembl = df_embedded_drugs_chembl[df_embedded_drugs_chembl['hc_order'] == 8].copy()



df_2_canvass_natprod = df_embedded_canvass_natprod[df_embedded_canvass_natprod['hc_order'] == 2].copy()
df_3_canvass_natprod = df_embedded_canvass_natprod[df_embedded_canvass_natprod['hc_order'] == 3].copy()
df_4_canvass_natprod = df_embedded_canvass_natprod[df_embedded_canvass_natprod['hc_order'] == 4].copy()
df_5_canvass_natprod = df_embedded_canvass_natprod[df_embedded_canvass_natprod['hc_order'] == 5].copy()


df_2_drugs_natprod = df_embedded_drugs_natprod[df_embedded_drugs_natprod['hc_order'] == 2].copy()
df_3_drugs_natprod = df_embedded_drugs_natprod[df_embedded_drugs_natprod['hc_order'] == 3].copy()
df_4_drugs_natprod = df_embedded_drugs_natprod[df_embedded_drugs_natprod['hc_order'] == 4].copy()
df_5_drugs_natprod = df_embedded_drugs_natprod[df_embedded_drugs_natprod['hc_order'] == 5].copy()





```

```python
# Plotting
```

```python
# Compare HCASE ChEMBL Embeddings


fig = plt.figure()

fig, subfigures = plt.subplots(nrows = 4, ncols = 2, figsize = (15, 22))



panel_dual_plot (subfigures[0][0], df_2_canvass_chembl, df_2_drugs_chembl, subfig_label = 'z = 2')
panel_dual_plot (subfigures[0][1], df_3_canvass_chembl, df_3_drugs_chembl, subfig_label = 'z = 3')
panel_dual_plot (subfigures[1][0], df_4_canvass_chembl, df_4_drugs_chembl, subfig_label = 'z = 4')
panel_dual_plot (subfigures[1][1], df_5_canvass_chembl, df_5_drugs_chembl, subfig_label = 'z = 5')
panel_dual_plot (subfigures[2][0], df_5_canvass_chembl, df_5_drugs_chembl, subfig_label = 'z = 6')
panel_dual_plot (subfigures[2][1], df_5_canvass_chembl, df_5_drugs_chembl, subfig_label = 'z = 7')
panel_dual_plot (subfigures[3][0], df_5_canvass_chembl, df_5_drugs_chembl, subfig_label = 'z = 8')

subfigures[3][1].axis('off')





subfigures[0][0].text(-0.25, 1.25, "A", transform=subfigures[0][0].transAxes,
            fontsize=64, fontweight='bold', va='top')
subfigures[0][1].text(-0.25, 1.25, "B", transform=subfigures[0][1].transAxes,
            fontsize=64, fontweight='bold', va='top')
subfigures[1][0].text(-0.25, 1.25, "C", transform=subfigures[1][0].transAxes,
            fontsize=64, fontweight='bold', va='top')
subfigures[1][1].text(-0.25, 1.25, "D", transform=subfigures[1][1].transAxes,
            fontsize=64, fontweight='bold', va='top')
subfigures[2][0].text(-0.25, 1.25, "E", transform=subfigures[2][0].transAxes,
            fontsize=64, fontweight='bold', va='top')
subfigures[2][1].text(-0.25, 1.25, "F", transform=subfigures[2][1].transAxes,
            fontsize=64, fontweight='bold', va='top')
subfigures[3][0].text(-0.25, 1.25, "G", transform=subfigures[3][0].transAxes,
            fontsize=64, fontweight='bold', va='top')



fig.suptitle('Comparison of Embeddings in HCASE ChEMBL Space', fontsize = 32, y = 1)
#fig.supylabel('Dim 2')


plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.5)

plt.savefig (fname_out_compare_hcase_chembl_embedding_plot, dpi=300)

plt.show()


```

```python
# Compare HCASE NatProd Embeddings






fig = plt.figure()

fig, subfigures = plt.subplots(nrows = 2, ncols = 2, figsize = (14, 10))



panel_dual_plot (subfigures[0][0], df_2_canvass_natprod, df_2_drugs_natprod, subfig_label = 'z = 2')
panel_dual_plot (subfigures[0][1], df_3_canvass_natprod, df_3_drugs_natprod, subfig_label = 'z = 3')
panel_dual_plot (subfigures[1][0], df_4_canvass_natprod, df_4_drugs_natprod, subfig_label = 'z = 4')
panel_dual_plot (subfigures[1][1], df_5_canvass_natprod, df_5_drugs_natprod, subfig_label = 'z = 5')





subfigures[0][0].text(-0.25, 1.25, "A", transform=subfigures[0][0].transAxes,
            fontsize=64, fontweight='bold', va='top')
subfigures[0][1].text(-0.25, 1.25, "B", transform=subfigures[0][1].transAxes,
            fontsize=64, fontweight='bold', va='top')
subfigures[1][0].text(-0.25, 1.25, "C", transform=subfigures[1][0].transAxes,
            fontsize=64, fontweight='bold', va='top')
subfigures[1][1].text(-0.25, 1.25, "D", transform=subfigures[1][1].transAxes,
            fontsize=64, fontweight='bold', va='top')





fig.suptitle('Comparison of Embeddings in HCASE NatProd Space', fontsize = 32, y = 1.05)
#fig.supylabel('Dim 2')


plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

plt.savefig (fname_out_compare_hcase_natprod_embedding_plot, dpi=300)

plt.show()





```
